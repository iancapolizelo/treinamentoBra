```
import numpy as np
```

### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
arr = np.random.randint(1, 51, size=(10))
print('Array original: ', arr)

media = np.median(arr)
print('Média: ', media)

desvio = np.std(arr)
print('Desvio padrão: ', desvio)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
arr = np.random.rand(10)
print('Array original: ', arr)

arr[arr <= 0.5] = np.sqrt(arr[arr <= 0.5])
print('Array com raiz quadrada: ', arr)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz = np.random.rand(3, 3, 3)
print('Matriz original: ', matriz)

matriz_2d = np.max(matriz, axis=1)

print('Matriz 2D: ', matriz_2d)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz = np.random.rand(5,5)
print('Matriz original: ', matriz)

media = np.median(matriz, axis=1)
print('Media: ', media)

matriz_nova = matriz - media[:, np.newaxis]
print('Matriz nova: ', matriz_nova)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4,4)
print('Matriz original: ', matriz)

diagonal_principal = np.diag(matriz)
print('Diagonal principal: ', diagonal_principal)

diagonal_secundaria = np.diag(np.fliplr(matriz))
print('Diagonal secundaria: ', diagonal_secundaria)

media_principal = np.mean(diagonal_principal)
media_secundaria = np.mean(diagonal_secundaria)

print('Média da diagonal principal: ', media_principal)
print('Média da diagonal secundaria: ', media_secundaria)

desvio_principal = np.std(diagonal_principal)
desvio_secundaria = np.std(diagonal_secundaria)

print('Desvio padrão principal: ', desvio_principal)
print('Desvio padrão secundaria: ', desvio_secundaria)
```
